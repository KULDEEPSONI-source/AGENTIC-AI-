# Introduction to Agentic AI

## What Is Agentic AI?

**Agentic AI** refers to artificial intelligence systems that can act as **autonomous agents** — they don't just generate text or predictions, they can **plan, reason, make decisions, and take actions** to achieve goals.

Unlike traditional AI (which responds to single prompts), agentic AI operates in **loops**: observe → think → act → repeat.

---

## The Core Loop: ReAct Pattern

The most common framework is **ReAct** (Reasoning + Acting):

```
┌─────────┐    ┌──────────┐    ┌────────┐    ┌─────────┐
│ Thought │───→│  Action  │───→│Observe │───→│ Thought │───→ ...
│(Reason) │    │(Tool Use)│    │(Result)│    │(Reason) │
└─────────┘    └──────────┘    └────────┘    └─────────┘
```

**Example in action:**
1. **Thought**: "I need to find the weather in Tokyo"
2. **Action**: Call weather API with "Tokyo"
3. **Observation**: API returns "25°C, sunny"
4. **Thought**: "Now I can answer the user's question"

---

## Key Components

| Component | What It Does | Example |
|-----------|-------------|---------|
| **Planning** | Breaks goals into sub-tasks | "Research topic → Outline → Write → Edit" |
| **Memory** | Stores context across interactions | Conversation history, learned facts |
| **Tool Use** | Interacts with external systems | APIs, databases, code execution |
| **Reasoning** | Makes decisions about what to do next | "Should I search or calculate?" |
| **Reflection** | Evaluates and corrects mistakes | "My last action failed, try differently" |

---

## Architecture Patterns

### 1. Single Agent
One agent with tools handles everything.

```
User → [Agent + Tools] → Result
```

### 2. Multi-Agent Systems
Specialized agents collaborate:

```
┌─────────┐     ┌─────────┐     ┌─────────┐
│Researcher│────→│ Writer  │────→│ Editor  │
│  Agent   │     │  Agent  │     │  Agent  │
└─────────┘     └─────────┘     └─────────┘
```

### 3. Hierarchical
A "manager" agent delegates to worker agents.

## Simple Example (Python Concept)

```python
# Conceptual agent loop
def agent_run(goal, tools, max_steps=10):
    memory = []
    for step in range(max_steps):
        # 1. Think
        thought = llm.reason(goal, memory)
        
        # 2. Act
        action = llm.choose_tool(thought, tools)
        result = execute(action)
        
        # 3. Remember
        memory.append({"thought": thought, "action": action, "result": result})
        
        # 4. Check if done
        if is_complete(goal, result):
            return result
    
    return "Max steps reached"
```

# What Are "Tools" in Agentic AI?

In agentic AI, **tools** are **functions or capabilities** that an agent can call to interact with the **outside world** — beyond just generating text.

Think of it this way: the LLM is the "brain," and tools are its **hands, eyes, and voice** to the real world.

---

## The Core Idea

```
┌─────────┐      Tool Call      ┌─────────┐
│   LLM   │ ──────────────────→ │  Tool   │
│  Brain  │  "search_web('AI')" │  Function│
│         │                     │         │
│         │ ←────────────────── │         │
│         │     Result          │         │
│         │  "AI means..."      │         │
└─────────┘                     └─────────┘
```

The LLM **decides** when to use a tool, **calls** it with parameters, and **uses** the result in its reasoning.

---

## Common Tool Categories

### 1. **Information Tools** (Eyes)
| Tool | What It Does | Example Call |
|------|-------------|--------------|
| **Web Search** | Searches the internet | `search("latest AI news 2026")` |
| **Database Query** | Reads from databases | `query("SELECT * FROM users WHERE id=5")` |
| **File Reader** | Reads local files | `read_file("report.pdf")` |
| **API Call** | Hits external APIs | `get_weather("Tokyo")` |

### 2. **Action Tools** (Hands)
| Tool | What It Does | Example Call |
|------|-------------|--------------|
| **Send Email** | Sends messages | `send_email(to, subject, body)` |
| **Book Calendar** | Creates events | `create_event("Meeting", "10am")` |
| **Write File** | Saves to disk | `write_file("output.txt", "data")` |
| **Run Code** | Executes Python/JS | `execute_python("2+2")` |
| **Deploy** | Pushes code live | `deploy_app("v1.2.3")` |

### 3. **Calculation Tools** (Brain Boosters)
| Tool | What It Does | Example Call |
|------|-------------|--------------|
| **Calculator** | Math operations | `calculate("345 * 782")` |
| **Code Execution** | Complex logic | `run_code("import pandas; df.mean()")` |

---

## How It Works in Practice

### Example: Agent answering "What's the weather in Paris?"

**Step 1: LLM decides it needs a tool**
```
LLM thinks: "I don't know current weather. I have a 'get_weather' tool. I'll use it."
```

**Step 2: LLM generates a tool call**
```json
{
  "tool": "get_weather",
  "parameters": {
    "location": "Paris"
  }
}
```

**Step 3: Tool executes**
```python
# Behind the scenes
def get_weather(location):
    return api_call(f"https://weather.api/{location}")
# Returns: {"temp": 22, "condition": "Sunny"}
```

**Step 4: LLM uses result to answer**
```
LLM sees: {"temp": 22, "condition": "Sunny"}
LLM responds: "It's 22°C and sunny in Paris right now!"
```

---

## Tool Definition Format (OpenAI Style)

When you give tools to an agent, you define them as **schemas**:

```python
tools = [
    {
        "type": "function",
        "function": {
            "name": "get_weather",
            "description": "Get current weather for a location",
            "parameters": {
                "type": "object",
                "properties": {
                    "location": {
                        "type": "string",
                        "description": "City name, e.g. 'Paris'"
                    }
                },
                "required": ["location"]
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "send_email",
            "description": "Send an email to a user",
            "parameters": {
                "type": "object",
                "properties": {
                    "to": {"type": "string"},
                    "subject": {"type": "string"},
                    "body": {"type": "string"}
                },
                "required": ["to", "subject", "body"]
            }
        }
    }
]
```

The LLM uses the **description** to decide *which* tool to use, and the **parameters** schema to know *what* to pass.

## Structured Tool Use (Modern Approach)

```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain.agents import create_tool_calling_agent, AgentExecutor
from langchain import hub

# Get pre-built tools
from langchain_community.tools import DuckDuckGoSearchRun, WikipediaQueryRun

# Custom tool
@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers."""
    return a * b

# Combine all tools
tools = [
    DuckDuckGoSearchRun(),
    WikipediaQueryRun(),
    multiply
]

# Create agent
llm = ChatOpenAI(model="gpt-4o")
prompt = hub.pull("hwchase17/openai-functions-agent")
agent = create_tool_calling_agent(llm, tools, prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

# Run
result = agent_executor.invoke({"input": "What is 42 × 7 and who invented the telephone?"})
```

# Deep Dive: LangChain Tools — All Patterns Explained

## 1. Custom Tools Using Decorators

### Basic `@tool` Decorator

```python
from langchain_core.tools import tool

# Simple tool - LangChain auto-detects name, description, and schema
@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers. Returns the product of a and b."""
    return a * b

# The tool is now a callable with metadata
print(multiply.name)        # "multiply"
print(multiply.description) # "Multiply two numbers..."
print(multiply.args)        # {"a": {"type": "int"}, "b": {"type": "int"}}

# Use it
result = multiply.invoke({"a": 5, "b": 3})
print(result)  # 15
```

### Decorator with Custom Name

```python
@tool("calculator_multiply")
def multiply(a: int, b: int) -> int:
    """Multiply two numbers."""
    return a * b

print(multiply.name)  # "calculator_multiply"
```

### Decorator with `parse_docstring` (Auto-schema from Google-style docstrings)

```python
@tool(parse_docstring=True)
def send_email(
    to: str,
    subject: str,
    body: str,
    priority: str = "normal"
) -> str:
    """
    Send an email to a recipient.

    Args:
        to: Email address of the recipient
        subject: Subject line of the email
        body: Main content of the email
        priority: Priority level (low, normal, high)
    """
    # Implementation
    return f"Email sent to {to}"

# Schema auto-generated from docstring
print(send_email.args)
# {
#   "to": {"description": "Email address...", "type": "string"},
#   "subject": {"description": "Subject line...", "type": "string"},
#   ...
# }
```

## 2. Structured Tool (Pydantic-Based)

When you need **complex validation**, **nested schemas**, or **field descriptions**, use Pydantic models.

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field

# Define input schema
class WeatherInput(BaseModel):
    location: str = Field(description="City name, e.g. 'Paris'")
    units: str = Field(default="celsius", description="Temperature units: 'celsius' or 'fahrenheit'")
```

## Safe Usage Patterns

### 1. **Read-Only Mode** (Recommended)

```python
from langchain_community.tools import ShellTool

class ReadOnlyShellTool(ShellTool):
    """Shell tool that only allows safe read commands."""
    
    # Override to block dangerous commands
    def _run(self, commands: str) -> str:
        # Block list
        dangerous = ['rm', 'mv', 'cp', 'sudo', '>', '>>', 'curl', 'wget', 'ssh']
        
        for cmd in dangerous:
            if cmd in commands.lower():
                return f"ERROR: Command '{cmd}' is not allowed for safety."
        
        return super()._run(commands)

safe_shell = ReadOnlyShellTool()
```

### 2. **Whitelist Approach**

```python
import subprocess
from langchain_core.tools import tool

ALLOWED_COMMANDS = {
    'ls': 'ls',
    'pwd': 'pwd',
    'cat': 'cat',
    'echo': 'echo',
    'date': 'date',
    'whoami': 'whoami',
    'ps': 'ps aux',
    'df': 'df -h',
    'du': 'du -sh',
    'grep': 'grep',
    'find': 'find',
    'head': 'head',
    'tail': 'tail',
    'wc': 'wc',
    'sort': 'sort',
    'uniq': 'uniq'
}

@tool
def safe_shell(command: str) -> str:
    """
    Execute a safe shell command.
    Only whitelisted commands are allowed.
    """
    # Extract base command
    base = command.strip().split()[0]
    
    if base not in ALLOWED_COMMANDS:
        return f"ERROR: Command '{base}' is not in the whitelist. Allowed: {list(ALLOWED_COMMANDS.keys())}"
    
    try:
        result = subprocess.run(
            command,
            shell=True,
            capture_output=True,
            text=True,
            timeout=10,
            cwd="/tmp"  # Restrict to safe directory
        )
        return result.stdout or result.stderr
    except subprocess.TimeoutExpired:
        return "ERROR: Command timed out after 10 seconds"
    except Exception as e:
        return f"ERROR: {str(e)}"

# Use safe_shell instead of ShellTool
```

The code snippet `multiply.args_schema.model_json_schema()` is used in **LangChain** to retrieve the **JSON Schema representation** of the input arguments required by a tool named `multiply`.

### What This Does

In LangChain, tools are often defined with an `args_schema` (typically a Pydantic `BaseModel`) that dictates what inputs the tool expects. Calling `.model_json_schema()` on this schema generates a standard JSON dictionary that describes these inputs.

This is critical because it allows the **LLM (Large Language Model) to understand** exactly how to call the tool. The LLM reads this JSON schema to determine:

1. **Which parameters are required** (e.g., `a` and `b`).
2. **What data types those parameters must be** (e.g., `integer` or `float`).
3. **Descriptions of the parameters**, which help the LLM know what values to provide.

### Example Context

If you have a `multiply` tool defined like this:

```python
from langchain.tools import tool
from pydantic import BaseModel, Field

class MultiplyInput(BaseModel):
    a: int = Field(description="The first number")
    b: int = Field(description="The second number")

@tool("multiply", args_schema=MultiplyInput)
def multiply(a: int, b: int) -> int:
    """Multiply two numbers."""
    return a * b

```

When you call `multiply.args_schema.model_json_schema()`, the output will look something like this:

```json
{
  "title": "MultiplyInput",
  "type": "object",
  "properties": {
    "a": {
      "title": "A",
      "description": "The first number",
      "type": "integer"
    },
    "b": {
      "title": "B",
      "description": "The second number",
      "type": "integer"
    }
  },
  "required": ["a", "b"]
}

```

### Why It Matters

* **Agent Interaction:** Agents use this schema to construct the correct function calls. Without this schema, the agent wouldn't know how to structure the data to pass to the `multiply` function.
* **Validation:** Pydantic uses this same schema to validate that the input provided by the LLM matches the expected format before the tool code actually executes.
* **Tool Documentation:** It provides a machine-readable way to document the interface of your tools.

## **Custom Tools (Method 2)**
#### **StructuredTool & Pydantic**

To understand how to create tools in LangChain, think of **`StructuredTool`** and **`BaseModel`** as a set of instructions that teach an AI how to use a specific piece of software.

Without these, the AI just sees a block of code; with them, the AI understands exactly what inputs are needed and how to "press the buttons" to get a result.

---

### 1. `BaseModel` (The "Recipe Card")

Think of `BaseModel` as a **standardized form or recipe card**. If you want an AI to perform a task like "Calculate Tax," it needs to know exactly what numbers to use.

By creating a `BaseModel`, you are defining the structure of the information you need. You are essentially telling the AI: *"To do this job, I need a piece of paper that must have a 'Price' field (a number) and a 'TaxRate' field (a number)."*

* **Why use it?** It forces the AI to provide data in a format your code can actually read. If the AI tries to give you a word when it should be a number, `BaseModel` catches the error before the code even runs.

### 2. `Field` (The "Label Maker")

`Field` works right inside your `BaseModel`. If `BaseModel` is the recipe card, **`Field` is the label or the instruction manual** you write on the fields of that card.

Instead of just saying "I need a number," you use `Field` to say: *"I need a number, and by the way, this is the 'Price,' and it should be a positive number."*

* **Why use it?** The AI isn't a mind reader. The description you put inside `Field` is the main way you "explain" to the AI what that specific input does.

### 3. `StructuredTool` (The "Interface")

`StructuredTool` is the **bridge** that connects your "recipe card" (the schema) to your actual Python function. It wraps your code into a package that LangChain can easily show to an AI agent.

When you create a `StructuredTool`, you are telling LangChain: *"Here is a function that does the math, and here is the `BaseModel` that explains how to use it. Now, go tell the AI how it works."*

---

### Putting it all together: A Simple Example

Imagine we want to build a tool that calculates the total price of an item including tax.

```python
from langchain.tools import StructuredTool
from pydantic import BaseModel, Field

# 1. The Recipe Card (BaseModel)
class CalculateTotalSchema(BaseModel):
    price: float = Field(description="The original price of the item.")
    tax_rate: float = Field(description="The tax rate as a decimal (e.g., 0.05 for 5%).")

# 2. The Logic
def calculate_total(price: float, tax_rate: float):
    return price + (price * tax_rate)

# 3. The Bridge (StructuredTool)
tool = StructuredTool.from_function(
    func=calculate_total,
    name="calculate_total",
    description="Calculates the total price including tax.",
    args_schema=CalculateTotalSchema
)

```

In this setup:

* The **LLM** reads the `description` inside the `StructuredTool` to decide *if* it should use the tool.
* The **LLM** reads the `BaseModel` and `Field` descriptions to decide *how* to fill in the `price` and `tax_rate`.
* **`StructuredTool`** ensures the final numbers are sent safely to the `calculate_total` function.

````python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
````

````python
class MultiplyInput(BaseModel):
    a: int = Field(description="The first number")
    b: int = Field(description="The second number")
````

In your code, `BaseModel` acts as the **blueprint, validator, and translator** for the data your tool needs.

Think of it as the "Contract" between your code and the AI. Here is the specific role it plays:

### 1. It Defines the "Schema" (The Blueprint)

The AI needs to know exactly what kind of information you are looking for. By inheriting from `BaseModel`, you are telling LangChain: *"My tool requires a specific object that contains exactly two things: an integer named `a` and an integer named `b`."* Without this, the AI might try to send text, a list, or nothing at all.

### 2. It Provides Automatic Validation (The Guardrail)

`BaseModel` is the security guard. Before your `multiply` function ever runs, the `BaseModel` inspects the data the AI sent:

* **Did the AI send numbers?** If the AI sends the word "five," `BaseModel` will block it and throw an error, preventing your code from crashing.
* **Are all required fields present?** If the AI only sends `a` but forgets `b`, `BaseModel` identifies the missing information immediately.

### 3. It Acts as a Translator (The API Interface)

Large Language Models (LLMs) do not "understand" Python functions. They operate by generating text. `BaseModel` allows LangChain to convert your Python requirements into a **JSON Schema** (as you saw in your first question).

* The LLM reads that JSON schema to understand the "rules" of your tool.
* When the LLM decides to use the tool, it generates a JSON object that perfectly matches the `BaseModel` you defined.

---

### Visualize the Flow

You can imagine the `BaseModel` as a filter that the AI must pass through to reach your Python logic.

> **Note:** Because you are using `BaseModel`, you get **type safety** for free. You don't have to write manual `if` statements to check if `a` is an integer; Pydantic does that for you automatically.

Does understanding that `BaseModel` is essentially a "contract" help clarify why it's necessary for your code to work with an AI?

## **Custom Tools (Method 3)**
#### **BaseTool**

````python
from langchain.tools import BaseTool
from typing import Type
````

````python
# arg schema using pydantic

class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to add")
    b: int = Field(required=True, description="The second number to add")
````

````python
class MultiplyTool(BaseTool):
    name: str = "multiply"
    description: str = "Multiply two numbers"

    args_schema: Type[BaseModel] = MultiplyInput

    def _run(self, a: int, b: int) -> int:
        return a * b
````

````python
multiply_tool = MultiplyTool()
````

````python
result = multiply_tool.invoke({'a':3, 'b':5})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)

print(multiply_tool.args)
````

## Toolkit

````python
from langchain_core.tools import tool
````

````python
# Custom tool
@tool
def add(a: int, b: int) -> int:
  """Add two numbers"""
  return a+b

@tool
def multiply(a: int, b: int) -> int:
  """Multiply two numbers"""
  return a*b
````

````python
class MathToolKit:
  def get_tools(self):
    # 'add' and 'multiply' are globally defined tool objects from cell 1cPwuoZISimh
    return [add, multiply]
````

````python
toolkit = MathToolKit()
tools = toolkit.get_tools()

for tool in tools:
  print(tool.name, "->", tool.description)
````

Now that the `add` and `multiply` tools are defined and accessible through the `MathToolKit`, let's demonstrate how to use them by passing arguments.

````python
# Use the 'add' tool to add two numbers
add_result = add.invoke({"a": 10, "b": 5})
print(f"Result of adding 10 and 5: {add_result}")
````

```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
import requests
```

```python
# Create a tool

@tool

def multiply(a: int, b:int) -> int:
  """Given 2 numbers a and b this tool returns their product"""
  return a*b
```

```python
print(multiply.invoke({'a':3, 'b':5}))
```

```python
multiply.name
```

```python
multiply.description
```

```python
multiply.args
```

```python
# Tool binding

llm = ChatOpenAI()
```

```python
# Bind multiple tools
# llm.bind_tools([tool1, tool2])
```

```python
llm.invoke('hi')
```

```python
llm_with_tools = llm.bind_tools([multiply])
```

```python
llm_with_tools.invoke('hi how are you')
```

```python
llm_with_tools.invoke('can you multiply 3 with 10')
```

```python
result = llm_with_tools.invoke('can you multiply 3 with 10')
```

```python
# Tool Creation: @tool, StructuredTool, BaseTool
# Tool Binding
# Tool Calling
# Tool Execution
```
